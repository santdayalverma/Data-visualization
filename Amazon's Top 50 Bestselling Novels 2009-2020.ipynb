```python
path = "/content/gdrive/MyDrive/Datasets/Amazon's Top 50 Bestselling Novels 2009-2020/AmazonBooks - Sheet1.csv"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv(path) 
df.head()
```

#Data Visualization

```python
#Genre
plt.figure(figsize=(15,5))
sns.set_theme(style="darkgrid")
sns.countplot(x="Genre", data=df)
plt.ylabel(ylabel="Number of Books")
```

```python
#User Rating
plt.figure(figsize=(20,5))
#sns.histplot(x=df["User Rating"]) "both are same code"
sns.histplot(x="User Rating", data = df, fill=True, kde=True)
plt.title("User Rating distribution")
```

```python
#User Rating
plt.figure(figsize=(20,5))
g=sns.displot(x="User Rating", data = df, hue="Genre", fill=True, kind="kde")
plt.title("User Rating distribution")
g.fig.set_figwidth(20)
g.fig.set_figheight(5)
```

```python
plt.figure(figsize=(20,8))
sns.boxenplot(x="Year", y="User Rating", data=df)
plt.title('User rating distribution by year')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.pointplot(x="Year", y="User Rating", hue="Genre", data=df)
```

```python
plt.figure(figsize=(20,5))
sns.histplot(x="Reviews", data=df, fill=True, kde=True)
```

```python
g=sns.displot(x="Reviews", data=df, hue="Genre", fill= True, kde=True)
g.fig.set_figwidth(20)
g.fig.set_figheight(5)
```

```python
df.query("Reviews > 9000")
```

```python
plt.figure(figsize=(20,5))
sns.pointplot(x="Year", y="Reviews", hue="Genre", data=df, apha= .3)
```

```python
plt.figure(figsize=(20,5))
sns.boxenplot(x="Year", y="Reviews", data=df)
```

```python
from wordcloud import WordCloud

# Generate a word cloud image
wordcloud = WordCloud(background_color='white').generate(str(df['Author'].values))

plt.figure(figsize=(40, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.title('Main authors')
plt.show()
```

#Combination

```python
fig, ax = plt.subplots(1, 2, figsize=(22, 10))
sns.scatterplot(x="Price", y="Reviews", hue="Genre", data=df, ax=ax[0])
sns.scatterplot(x="Price", y="User Rating", hue="Genre", data=df, ax=ax[1])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(22, 10))
sns.regplot(x="Year", y="Reviews", data=df, ax=ax[0])
sns.regplot(x="Year", y="Price", data=df, ax=ax[1])
```
