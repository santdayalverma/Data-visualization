```python
data = '/content/drive/MyDrive/Kaggle_Datasets/Agriculture crop production/datafile (2).csv'
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_2 = pd.read_csv(data)
```

```python
df_2.head()
```

```python
df_2.columns = ['Crop', 'Production 2006-07', 'Production 2007-08',
       'Production 2008-09', 'Production 2009-10', 'Production 2010-11',
       'Area 2006-07', 'Area 2007-08', 'Area 2008-09', 'Area 2009-10',
       'Area 2010-11', 'Yield 2006-07', 'Yield 2007-08', 'Yield 2008-09',
       'Yield 2009-10', 'Yield 2010-11']
```

```python
#2006-07
plt.figure(figsize=(20,7))
sns.lineplot(x="Crop",y="Production 2006-07", data=df_2 ,marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x="Crop",y="Area 2006-07", data=df_2, marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df_2,x="Crop",y="Yield 2006-07",marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
# comparison between 

plt.figure(figsize=(20,7))
sns.barplot(x="Crop",y="Production 2006-07", data=df_2, color='lightblue')
sns.lineplot(x="Crop",y="Area 2006-07", data=df_2, marker='o',label='Area 2006-07')
sns.lineplot(x="Crop",y="Yield 2006-07", data=df_2, marker='o',label='Yield 2006-07')
plt.xticks(rotation=90)
```
